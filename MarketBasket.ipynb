```python
import pandas as pd
import numpy as np
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from pandas.plotting import parallel_coordinates
from itertools import permutations
```

```python
data=pd.read_csv("teleco_market_basket.csv")
data.info()
```

```python
#remove blank rows
data=data.dropna(axis=0, how='all')
data.head()
```

```python
#remove blank columns
data=data.dropna(axis=1, how='all')
data=data.fillna('None')
data.head()
```

```python
#create list of lists
lists=data.values.tolist()
```

```python
#fit transaction encoder
encoder=TransactionEncoder()
encoder.fit(lists)

lists_array=encoder.transform(lists)
lists_df=pd.DataFrame(lists_array, columns=encoder.columns_)
lists_df
```

```python
#remove blank columns
lists_df=lists_df.dropna(axis=1)
lists_df
```

```python
pd.DataFrame(lists_df).to_csv("cleaned_task3.csv", index=False)
```

```python
#frequent itemsets
frequent=apriori(lists_df, min_support=0.02, use_colnames=True)
frequent.head(5)
```

```python
#association rules
rules=association_rules(frequent, metric='lift', min_threshold=1)
rules.head()
```

```python
#sort by metric
top_three=rules.sort_values('confidence', ascending=False).head(3)
top_three
```

```python
top_three=rules.sort_values('lift', ascending=False).head(3)
top_three
```

```python
top_three=rules.sort_values('support', ascending=False).head(3)
top_three
```

```python
sorted_rules=rules[(rules['lift']>0.08)]
sorted_rules.head(3)
```
